```python
# Eliminar columnas que no aportan valor predictivo
df = df.drop(columns=['CustomerID'])
```

```python
# Unificar "No internet service" como "No" para evitar categorías redundantes antes del encoding.
cols_to_fix = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in cols_to_fix:
    df[col] = df[col].replace({'No internet service': 'No'})
```

```python
# Validar la presencia de datos faltantes antes del modelado
df.isnull().sum()
```

```python
# Calcular y visualizar la matriz de correlación entre variables numéricas y Churn.
df_numeric = df.select_dtypes(include=["int64", "float64"])

df_numeric["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

corr_matrix = df_numeric.corr()

# Visualizar heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación - Variábles Numericas")
plt.show()
```

```python
#Tiempo (Tenure) vs Churn — Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Churn"], y=df["Tenure"])
plt.title("Tenure (tiempo) vs Cancelación (Churn)")
plt.xlabel("Churn")
plt.ylabel("Tenure")
plt.show()
```

```python
#“Gasto total” vs Churn — Boxplot
df["TotalEstimado"] = df["ChargesMonthly"] * df["Tenure"]

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Churn"], y=df["TotalEstimado"])
plt.title("Gasto total estimado vs Cancelación (Churn)")
plt.xlabel("Churn")
plt.ylabel("TotalEstimado")
plt.show()
```

```python
# Eliminar variables redundantes para reducir multicolinealidad y evitar información duplicada.
columns_to_drop = [col for col in ["ChargesDaily","ChargesTotal"] if col in df.columns]
df = df.drop(columns=columns_to_drop)
```

```python
# Identificar variables categóricas excluyendo la variable objetivo (Churn).
categorical_cols = df.select_dtypes(include=["object"]).columns
categorical_cols = categorical_cols.drop("Churn")
```

```python
# Analizar la relación porcentual entre cada variable categórica y la cancelación (Churn).
for col in categorical_cols:
    print(f"\n📌 Analisis de variable: {col}")
    display(pd.crosstab(df[col], df["Churn"], normalize="index") * 100)
```

```python
# Aplicar el test Chi-cuadrado para evaluar la relación entre variables categóricas y la cancelación (Churn) y generar un ranking de importancia.
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Crea copia
df_rank = df.copy()

# Convertir Churn para binario
df_rank["Churn"] = df_rank["Churn"].map({"Yes": 1, "No": 0})

# Separar variables categóricas
categorical_cols = df_rank.select_dtypes(include=["object"]).columns

# Aplicar Label Encoding temporario
le = LabelEncoder()
for col in categorical_cols:
    df_rank[col] = le.fit_transform(df_rank[col])

# Separar X e y
X_cat = df_rank[categorical_cols]
y = df_rank["Churn"]

# Aplicar teste Qui-cuadrado
chi_scores, p_values = chi2(X_cat, y)

# Criar DataFrame com ranking
chi_df = pd.DataFrame({
    "Variable": categorical_cols,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})

# Ordenar por importancia
chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

chi_df
```

```python
significant_vars = chi_df[chi_df["p-value"] < 0.05]

print("\n✅ Variables significativas (p < 0.05):")
display(significant_vars)
```

```python
columns_to_drop = [col for col in ["Gender", "PhoneService"] if col in df.columns]
df = df.drop(columns=columns_to_drop)
```

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# Aplicar One-Hot Encoding en treno
X_train = pd.get_dummies(X_train, drop_first=True)

# Aplicar en teste
X_test = pd.get_dummies(X_test, drop_first=True)

# Garantir que treno y teste tengan las mismas columnas
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
y_train.value_counts(normalize=True) * 100
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Escalar datos (importante para Regresión Logística)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo con balanceo
log_model = LogisticRegression(max_iter=3000, class_weight="balanced", random_state=42)

log_model.fit(X_train_scaled, y_train)

y_pred_log = log_model.predict(X_test_scaled)

print("Regresión logística (con escalado y balanceo)")
print(classification_report(y_test, y_pred_log))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

rf_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced",
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest (balanced)")
print(classification_report(y_test, y_pred_rf))

# Matriz de confusión (gráfico)
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"], yticklabels=["No", "Yes"])
plt.title("Matriz de Confusión - Random Forest")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
#Comparación visual Logística vs Random Forest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

metrics = []

# Logistic Regression
metrics.append({
    "Modelo": "Logística",
    "Accuracy": accuracy_score(y_test, y_pred_log),
    "Precision": precision_score(y_test, y_pred_log, pos_label="Yes"),
    "Recall": recall_score(y_test, y_pred_log, pos_label="Yes"),
    "F1": f1_score(y_test, y_pred_log, pos_label="Yes")
})

# Random Forest
metrics.append({
    "Modelo": "Random Forest",
    "Accuracy": accuracy_score(y_test, y_pred_rf),
    "Precision": precision_score(y_test, y_pred_rf, pos_label="Yes"),
    "Recall": recall_score(y_test, y_pred_rf, pos_label="Yes"),
    "F1": f1_score(y_test, y_pred_rf, pos_label="Yes")
})

metrics_df = pd.DataFrame(metrics)
metrics_df
```

```python
# Gráfico Comparativo
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(7,4))

x = np.arange(len(metrics_df))

plt.bar(x - 0.2, metrics_df["Recall"], width=0.4, label="Recall")
plt.bar(x + 0.2, metrics_df["F1"], width=0.4, label="F1-score")

plt.xticks(x, metrics_df["Modelo"])
plt.ylabel("Valor")
plt.title("Comparación de desempeño: Logística vs Random Forest")
plt.legend()

plt.show()
```

### ***Análisis***
Random Forest presenta un mayor recall en la clase “Yes”, lo que indica una mejor capacidad para detectar clientes con riesgo de cancelación. Aunque la precisión puede variar ligeramente, el modelo ofrece un mejor equilibrio general según el F1-score. Dado que el objetivo del negocio es anticipar la cancelación, se prioriza el modelo con mayor recall.

# **Conclusión Final**
El objetivo del presente estudio fue desarrollar modelos predictivos capaces de anticipar la cancelación de clientes (churn) en TelecomX, mediante técnicas de aprendizaje supervisado y un proceso estructurado de preprocesamiento, análisis exploratorio y modelado.

Durante la fase de análisis, se identificaron variables con alta relación con la cancelación, destacándose el tipo de contrato, la antigüedad del cliente (Tenure), el método de pago y la contratación de servicios adicionales. En particular, los clientes con contratos mensuales y menor tiempo de permanencia presentaron mayor probabilidad de abandono.

Se implementaron y evaluaron dos modelos de clasificación:

1. Regresión Logística (con estandarización y balanceo de clases)

2. Random Forest (con balanceo de clases)

La evaluación se realizó utilizando las métricas Accuracy, Precision, Recall y F1-score, además del análisis mediante matriz de confusión.

Resultados obtenidos
1. Random Forest (balanced):

    a. Accuracy: 0.78

    b. Precision (Yes): 0.62

    c. Recall (Yes): 0.47

    d. F1-score (Yes): 0.53

2. Regresión Logística (escalada y balanceada):

    a. Accuracy: 0.74

    b. Precision (Yes): 0.51

    c. Recall (Yes): 0.79

    d. F1-score (Yes): 0.62



# **Análisis Comparativo**

Aunque el modelo Random Forest obtuvo mayor exactitud global (78% frente a 74%), su capacidad para identificar correctamente clientes que cancelan (Recall = 0.47) es considerablemente inferior a la Regresión Logística (Recall = 0.79).

En problemas de churn, el Recall en la clase positiva (“Yes”) es la métrica estratégicamente más relevante, ya que representa la proporción de clientes en riesgo correctamente detectados. Un bajo recall implica una alta cantidad de falsos negativos, es decir, clientes que cancelarán y no fueron identificados por el modelo.

En este sentido:

1. Random Forest prioriza la clasificación correcta de clientes activos.

2. Regresión Logística prioriza la detección de clientes en riesgo.

El F1-score también respalda esta conclusión, siendo superior en la Regresión Logística (0.62 frente a 0.53), lo que indica un mejor equilibrio entre precisión y recall en la clase de interés.
